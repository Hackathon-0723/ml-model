```python
transform = transforms.Compose([
                        PIL.Image.fromarray,
                        transforms.Resize([224, 224]),
                        transforms.ToTensor(),
                        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
                      ])
```

```python
dataset = data.WashHandDataset(root='dataset/sliced', cache='.cache/train.db', transform=transform)
label_num = len(consts.l2n)
embedding_size = 64
trunk_model = 'resnet34'
layer_sizes = [embedding_size, label_num]
classifier = model.Classifier(trunk_model, layer_sizes, final_relu=False)
```
